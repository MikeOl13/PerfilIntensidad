# Perfil de intensidad 

```python
# Se importan librerias

from matplotlib import pyplot as plt
import cv2
```

```python
# Asignamos una variable para la imagen
img = cv2.imread('../Fotos/B0.png',0)
  
# Buscamos la frecuencia de pixeles en un rango 0-255
histr = cv2.calcHist([img],[0],None,[256],[0,256])
  
# Mostramos la imagen
plt.plot(histr)
plt.show()
```
