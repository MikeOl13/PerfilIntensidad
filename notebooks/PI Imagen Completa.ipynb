# Perfil de intensidad 

En este documento se utilizan imagenes completas, sin segmentar.

## 1. Importamos librerias

```python
# Se importan librerias

from matplotlib import pyplot as plt
import cv2
```

```python
# Recorremos las imágenes desde B0 hasta B36
for i in range(37):
    # Formateamos el nombre de la imagen
    img_path = f'../intensidad/B{i}.png'
    
    # Leemos la imagen en modo gris
    img = cv2.imread(img_path, 0)
    
    # Verificamos si la imagen fue leída correctamente
    if img is None:
        print(f"No se pudo leer la imagen: {img_path}")
        continue
    
    # Calculamos el histograma de la imagen
    histr = cv2.calcHist([img], [0], None, [256], [0, 256])
    
    # Mostramos el histograma
    plt.plot(histr)
    plt.title(f'Histograma de B{i}.png')
    plt.xlabel('Intensidad de píxel')
    plt.ylabel('Frecuencia')
    plt.show()
```
